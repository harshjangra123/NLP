BAG of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Sample list of documents
corpus = [
    "The cat sat on the mat.",
    "The dog barked at the cat.",
    "Cats and dogs are great pets.",
    "I love my cat.",
    "My dog loves to play fetch.",
    "The cat and the dog are friends.",
    "Dogs are loyal and cats are independent."
]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
stopwords.words('english')
```

```python
for i in range(len(corpus)):
    words = word_tokenize(corpus[i])
    
    final_words = [] 
    for word in words:
        if word not in set(stopwords.words('english')):
            word = word.lower()
            word = lemmatizer.lemmatize(word)
            final_words.append(word)
    corpus[i] = ' '.join(final_words)
```

```python
vectorizer = CountVectorizer(lowercase=True,stop_words=stopwords.words('english'))
```

```python

# fit the vectorizer to the corpus and transform the corpus into document-term matrix
# the fit transform method learns the vocabulary and transform the text data into0 numerical array
```

```python
x = vectorizer.fit_transform(corpus)
```

```python
vectorizer.get_feature_names_out()
```

```python
print(x.toarray())
```

PCA in NLP

```python
import spacy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Sample list of documents
corpus = [
    "The cat sat on the mat.",
    "The dog barked at the cat.",
    "Cats and dogs are great pets.",
    "I love my cat.",
    "My dog loves to play fetch.",
    "The cat and the dog are friends.",
    "Dogs are loyal and cats are independent."
]
```

```python

sentences = []
for i in range(len(corpus)):
    words = word_tokenize(corpus[i])

    final_words = []
    for word in words:
        if word not in set(stopwords.words('english')) and word not in set(string.punctuation):
            word = word.lower()
            word = lemmatizer.lemmatize(word)
            final_words.append(word)
    sentences.append(final_words)
```

```python
sentences
```

```python
model = Word2Vec(sentences,vector_size=100,window=12,min_count=1,sg=0,workers=4)
```

```python
words = list(model.wv.index_to_key)
```

```python
word_vectors = model.wv[words]
```

```python
word_vectors.shape
```

```python
pca = PCA(n_components=2)
```

```python
word_vectors_pca = pca.fit_transform(word_vectors)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(word_vectors_pca[:,0],word_vectors_pca[:,1])
plt.grid(True)
```

cosine similarity and random projection LSH

```python
from sklearn.random_projection import SparseRandomProjection
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = Word2Vec(sentences,vector_size=3,window=12,min_count=1,sg=0,workers=4)
```

```python
words = list(model.wv.index_to_key)
```

```python
word_vectors = model.wv[words]
```

```python
word_vectors.shape
```

```python
word_vectors
```

```python
lsh = SparseRandomProjection(n_components=2)
dtata_lsh = lsh.fit_transform(word_vectors)
```

```python
cos_sim = cosine_similarity(dtata_lsh)
```

```python
cos_sim
```

```python
plt.scatter(dtata_lsh[:,0],dtata_lsh[:,1])
```
