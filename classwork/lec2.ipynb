```python
stemmer = PorterStemmer()
lemmetizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
nltk.download('wordnet')
```

```python
def preprocess(text):
    text = text.lower()

    # removong punctuations
    text = text.translate(str.maketrans('', '',string.punctuation))

    # tokenize
    tokens = text.split()

    tokens = [lemmetizer.lemmatize(stemmer.stem(word)) for word in tokens if word not in stop_words]

    return ' '.join(tokens)
```

```python
df['Review'] = df['Review'].apply(preprocess)
```

```python
df
```

```python
# vectorizer = TfidfTransformer()
# x = vectorizer.fit_transform(df['Review'])
# y = df['Sentiment']


from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the reviews into a TF-IDF matrix
x = vectorizer.fit_transform(df['Review'])

# Get the sentiment labels
y = df['Sentiment']

# Example: Print the shape of the resulting matrix
print(x.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
precision = precision_score(y_test, y_pred)
```
